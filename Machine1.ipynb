```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/insurance.csv")
df=data.head(10)
print(f"\nTop 10 data :",df)
df=data.tail(10)
print(f"\n Last ten data :",df)
df=data.info()
print(f"\nInformation :",df)
df=data.describe()
print(f"\nData describe :",df)
df=data.isnull().sum()
print(f"\nThe total null values :\n",df)
for col in data.columns:
  print(f"\nList of columns :\n",col)
num=["age",'children',"charges"]
for col in num:
  plt.figure(figsize=(9,9))
  sns.histplot(data[col],kde=True,bins=18)
plt.show()
print("\n======= HEATMAP =======\n")
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(numeric_only=True),annot=True)
plt.show()

print(f"\n=====Data is cleaned ======\n")
df=data.copy()
clean=df.head()
print(clean)

print(f"\n====Label Encoding =======\n")

df["sex"]=df["sex"].map({"female":1,"male":0})
df["smoker"]=df["smoker"].map({"yes":1,"no":0})
df["region"]=df["region"].map({"southwest":0,"northwest":1,"southeast":2,"northeast":3})

df.rename(columns={"sex":"is_female","smoker":"is_smoker"},inplace=True)
print(df.head())
print(df.tail())

print(f"\n0 means 'Southwest', 1 mean 'Northwest',2 means 'Southeast', 3 means 'Northeast'")
```
